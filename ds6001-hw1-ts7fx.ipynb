```python
# ds6001, hw1, tianye song, ts7fx, 10/20/2016
# Generate a 2d Gaussian plot
import numpy as np
import math
import matplotlib.pyplot as plt

# since this is a non-correlated case, use sd_x=1 & sd_y=1
sd_x,sd_y=1,1

# since used uniform spaced values for x&y, mu_x = 0 & mu_y = 0
points = np.arange(-5,5,0.01)
x, y = np.meshgrid(points, points)
mu_x,mu_y = 0,0

# plug in formula
z=1/np.sqrt(2*np.pi*sd_x*sd_y) * np.exp(-(((x-mu_x)**2/(2 * sd_x ** 2)) + ((y-mu_y)**2/(2 * sd_y ** 2))))

#draw the graph
%matplotlib inline

plt.imshow(z,cmap = plt.cm.gray); plt.colorbar()

plt.title("Image of 2d Gaussian")
```
